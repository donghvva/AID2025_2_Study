# Momentu, Adagrad, RMSProp, Adam, AdamW: SGD의 문제점을 개선한 옵티마이저

---



## 1. **Momentum**

---



### **핵심 아이디어**: 기울기 방향으로 갈 때 기울기를 그냥 쓰지 않고 이동 평균으로 모아서(관성) 그 방향으로 더 멀리간다. 즉, **이전의 기울기의 방향을 이용하여 다음 이동에 반영**하는 것

### **개선된 문제점**: 경사가 급하거나 축마다 경사가 다를 때 지그재그로 진동하며 느리게 내려감 →  흔들림(진동)이 줄고, 수렴 속도가 향상

## 2. **Adagrad**


---



### **핵심 아이디어**: 각 변수마다 기울기의 제곱 합을 누적해서, 많이 움직인 변수는 학습률을 자동으로 줄임

### **개선된 문제점**: 모든 가중치에 같은 학습률을 쓰면, 자주 변하는 변수는 너무 빨리 줄어듦 →  변수별로 적응적인 학습률 적용, 희소 데이터에 유리


## 3. **RMSProp**

---



### **핵심 아이디어**: 각 변수마다 기울기의 제곱 합의 누적 대신, 최근 기울기 제곱의 이동 평균(지수 감쇠) 을 사용해 **최신 정보만 반영**

### **개선된 문제점**: Adagrad는 시간이 지날수록 학습률이 0에 가까워져 멈춤 →  학습률 고갈 방지, 안정적인 수렴 유지

## 4. **Adam**

---



### **핵심 아이디어**: **기울기 평균(Momentum) + 기울기 제곱 평균(RMSProp) 을 함께 사용**하여 방향과 크기 모두 조절

### **개선된 문제점**: 학습 속도는 빠르게 하고 싶지만, 안정성도 필요함 → Momentum과 RMSProp 둘 다 써서 빠르고 안정적인 학습 가능

## 5. **AdamW**


---



### **핵심 아이디어**: Adam에서 가중치 감쇠를 경사 계산과 분리하여 정확히 적용

### 개선된 문제점: Adam은 가중치가 커져서 과적합이 생기거나, 정규화 효과가 약함 → 과적합 방지, 일반화 성능 향상 (딥러닝 최신 기본 설정)
